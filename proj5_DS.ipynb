## Introduction

Over the past fifteen years or so, "analytics" has grown into more of a prominent component of basketball player evaluation, especially in the world's most presteigous league, the National Basketball Association (NBA). Not only is it deployed by team front offices, but many of its fruits have become apart of the public discussion of the sport. Furthermore, more and more frequently, independent basketball junkies have been developing their own metrics. 

With this in mind, our research question/goal is to successfully measure a NBA player's ability to induce and convert offensive opportunities for himself, as well as leading teammates to those opportunities. In other words, we are attempting to measure how good a player is at creating and efficiently making shots for himself, and at making plays that lead to points for his teammates. We aim to create a statistic that is a solid representation of a player's self-induced offense, relative to those they are competing against. If our statistic is truly well-made, it could impact how teams and analysts evaluate this aspect of players' games.

Our new statistic is called "Self-Induced Offensive Measurement", or SIOM. However, because it is a standarized statistic, it is referred to as "rSIOM", where the "r" stands for "relative". rSIOM is a seasonal metric, dates back to the 2000-2001 season, and, for a given season, is only calculated for players in the top 70% of minutes played. The data used in our computations is all originally from Stathead, Basketball Reference, or pbpstats.

Before getting into more detail, here are some key terms which will be mentioned throughout the remainder of our report:

 - Offensive Box Plus/Minus (OBPM): an estimate of a player’s offensive contribution in points above league average per 100 possessions played that utilizes their box score information, position, and their team’s overall offensive performance.[<sup>1</sup>](#fn1)
 
 - True Shooting Percentage (TS%): a measure of shooting efficiency that takes into account field goals, 3-point field goals, and free throws.[<sup>2</sup>](#fn2)
 
 - Usage Rate (USG%): an estimate of the percentage of team plays used by a player while he was on the floor.[<sup>3</sup>](#fn3)
 
 - Pace: an estimate of the number of possessions per 48 minutes.[<sup>4</sup>](#fn4)
 
 - Offensive Rating (ORtg): for players it is points produced per 100 posessions, while for teams it is points scored per 100 possessions.[<sup>5</sup>](#fn5)

### I. Creating Our Statistic

```python
# Creating a new df to calculate and store the expected value of shots from 0-3 ft, 3-10 ft, 10-16 ft, 
# 16 ft to the 3PT line, corner 3s, and arc 3s for each individual season...
expected_values_by_loc = lg_avg_shooting.copy()
# Need to find % 3PA from arc (non-corner) and Arc 3PM %
expected_values_by_loc['pct_3pa_arc'] = 1 - expected_values_by_loc['pct_3pa_corner']
# Now, using the fact that overall 3PM % is a weighted average of arc and corner 3PM %, we know that...
expected_values_by_loc['pct_3pm_arc'] = round(((expected_values_by_loc['pct_3pm'] - 
                                        (expected_values_by_loc['pct_3pm_corner']*
                                         expected_values_by_loc['pct_3pa_corner']))/expected_values_by_loc['pct_3pa_arc']),3)
expected_values_by_loc.head(5)
```

In the above cell, we created a new DataFrame expected_values_by_loc which will contain the expected value of shots from various locations on the floor from 2000 to 2021. Furthermore, we created two columns which were not present in our read-in data `pct_3pa_arc` and `pct_3pm_arc`, which contains the league average percentage of 3s taken that were not in the corner and the conversion rate of non-corner 3s by season, respectively.

```python
# Now, finding the expected values...
expected_values_by_loc['ev_0to3ft'] = 2*(expected_values_by_loc['pct_fgm_0to3_ft'])
expected_values_by_loc['ev_3to10ft'] = 2*(expected_values_by_loc['pct_fgm_3to10_ft'])
expected_values_by_loc['ev_10to16ft'] = 2*(expected_values_by_loc['pct_fgm_10to16_ft'])
expected_values_by_loc['ev_16ftto3pt'] = 2*(expected_values_by_loc['pct_fgm_16ftto3pt'])
expected_values_by_loc['ev_3pt_corner'] = 3*(expected_values_by_loc['pct_3pm_corner'])
expected_values_by_loc['ev_3pt_arc'] = 3*(expected_values_by_loc['pct_3pm_arc'])
expected_values_by_loc

# Removing all unnecessary columns from expected_values_by_loc...
expected_values_by_loc = expected_values_by_loc.loc[:,['ev_0to3ft', 'ev_3to10ft', 'ev_10to16ft',
                                                       'ev_16ftto3pt', 'ev_3pt_corner', 'ev_3pt_arc']]
expected_values_by_loc.head(5)
```

In the above cell, we calculated and put into columns the expected value of shots from six different areas on the court for every season in our dataset (2000-2001 through 2020-2021) and then removed unnecessary columns from our DataFrame which contains the aforementioned expecrted values.

Note that pbpstats classifies short mid-range shots as shots from 4-14 ft away from the basket and long mid-range as coming from 14 ft away from the basket to the three point line. Because of how the Basketball Reference league average shooting data is broken up, we will have to define short mid-range as 3-16 ft away and long mid-range as 16 ft away to the three point line. Looking at the percentages in our lg_avg_shooting DataFrame, this re-classification should not alter our results significantly. In the below cell, we adjust our expected_values_by_loc DataFrame.

```python
# Adjusting for reclassification
for i in list(lg_avg_shooting.index.values):
    expected_values_by_loc.loc[i,'estev_3to16ft'] = round((((lg_avg_shooting.loc[i,'pct_fga_3to10_ft']/(lg_avg_shooting.loc[i,'pct_fga_3to10_ft'] 
                                                       + lg_avg_shooting.loc[i,'pct_fga_10to16_ft']))*expected_values_by_loc.loc[i,'ev_3to10ft']) 
                                                       + ((lg_avg_shooting.loc[i,'pct_fga_10to16_ft']/(lg_avg_shooting.loc[i,'pct_fga_3to10_ft'] 
                                                       + lg_avg_shooting.loc[i,'pct_fga_10to16_ft']))*expected_values_by_loc.loc[i,'ev_10to16ft'])),3)
expected_values_by_loc.head(5)
```

In the below cell, we find the `rSIOM` values for each player season present in our dataset. `rSIOM` is made up of two sub-metrics we've calculated, `siom_self_factor` and `siom_others_factor`. `siom_self_factor` is a measure of a player's self-induced scoring, while `siom_others_factor` is a measure of a player's playmaking. Both factors take into account both efficiency and production.

Additionally, after summing both factors, two further adjustments are made before getting our final unstandardized `siom` values. These adjustments are for player usage rate (`usg%`) and pace (`pace_estimate`). This is to prevent our statistic from favoring players who use a higher proportion of their team possesions while on the floor (which is measured by usage rate) or who have a larger number of possesions to produce points (which is measured by pace). This is because while production is significant, our statistic aims to measure a player's ability to induce offense, which is not the same as capturing a player's production while in that role.

```python
# Piece-by-piece building of our statistic
nba_groups = nba.groupby('season')
siom_self_factor = pd.Series([], dtype = float)
siom_others_factor = pd.Series([], dtype = float)
for name, group in nba_groups:
    siom_self_factor = siom_self_factor.append(((1 - (group['shotqualityavg']))*(group['efg%'] - lg_avg_teamstats.loc[name, 'efg%'])
                            *(group['ptsunassisted2s'] + group['ptsunassisted3s'])) + (group['ft%'] - 
                            (lg_avg_teamstats.loc[name, 'ft'])/lg_avg_teamstats.loc[name, 'fta'])*group['ftpoints'])
    siom_others_factor = siom_others_factor.append((1.5 - (group['tov%']/lg_avg_teamstats.loc[name, 'tov%']))
                                *((group['atrimassists']*(expected_values_by_loc.loc[name,'ev_0to3ft']))
                                + (group['shortmidrangeassists']*(expected_values_by_loc.loc[name,'estev_3to16ft']))
                                + (group['longmidrangeassists']*(expected_values_by_loc.loc[name,'ev_16ftto3pt']))
                                + (group['corner3assists']*(expected_values_by_loc.loc[name,'ev_3pt_corner']))
                                + (group['arc3assists']*(expected_values_by_loc.loc[name,'ev_3pt_arc']))))
nba['siom_self_factor'] = siom_self_factor
nba['siom_others_factor'] = (0.10018018018018018)*siom_others_factor
nba['siom_unadjusted'] = nba['siom_self_factor'] + nba['siom_others_factor']
# Adjusting for player USG%..
nba['uSIOM'] = (1 - np.log(nba['usg%']/20))*nba['siom_unadjusted']
# Adjusting for pace...
nba['pace_estimate'] = (48*nba['offposs'])/nba['mp']
nba_groups = nba.groupby('season')
unstded_SIOM = pd.Series([], dtype = float)
for name, group in nba_groups:
    unstded_SIOM = unstded_SIOM.append((1-np.log(group['pace_estimate']/lg_avg_teamstats.loc[name,'pace']))*group['uSIOM'])
nba['unstded_SIOM'] = unstded_SIOM
```

```python
nba.sort_values(by = 'unstded_SIOM', ascending = False)
```

Note that in the above output, there was no weight on either `siom_self_factor` or `siom_others_factor` when computing `unstded_SIOM`. By the output above, without performing further analysis, we can tell that `siom_others_factor` must have a disproportionate influence on any given player's `unstded_SIOM`. So, in the next two cells, we will attempt to find the optimal weight for `siom_others_factor` before standardizing our statistic by season. The new weight will be used in the cell where `unstded_SIOM` is calculated, however, the above output was obtained utilizing no weight on `siom_others_factor`.

```python
# Finding an optimal weight for "siom_others_factor"...
possible_coeffs = list(np.linspace(0.01,1,1000))
pct_influence = pd.DataFrame([pd.Series(np.zeros(1000)), pd.Series(np.zeros(1000))],
                             columns = ['siom_self', 'siom_others'],
                             dtype = float)
summed_siom_self_factor = sum(nba['siom_self_factor'])
for i in range(0,1000):
    alt_siom_others_factor = possible_coeffs[i]*nba['siom_others_factor']
    summed_alt_siom_others_factor = sum(alt_siom_others_factor)
    summed_unadj_siom = summed_siom_self_factor + summed_alt_siom_others_factor
    pct_influence.loc[i,'siom_self'] = summed_siom_self_factor/summed_unadj_siom
    pct_influence.loc[i,'siom_others'] = summed_alt_siom_others_factor/summed_unadj_siom
pct_influence
```

In the cell above, we found the percent influence of the sum of all the `siom_self_factor` and `siom_others_factor` values on the total unadjusted SIOM by multiplying the sum of all the `siom_others_factor` values by 1000 different possible coefficients before adding the two sums to retrive the resultant summed unadjusted SIOM.

Now, in the cell below, we will find the absolute difference between the percent influence of both factors and find a sensible weight to place on `siom_others_factor`.

```python
pct_influence['diff'] = abs(pct_influence['siom_self'] - pct_influence['siom_others'])
print(pct_influence.describe())
print(pct_influence[pct_influence['diff'] <= .6])
```

```python
print('Absolute percent influence difference with no weight: {:.4f}'.format(pct_influence.loc[999, 'diff']))
print('Absolute percent influence difference with selected coefficient({:.4f}):' 
      ' {:.4f}'.format(possible_coeffs[91], pct_influence.loc[91,'diff']))
```

After some trial and error, we settled on the above coefficient. This required some trial and error because the `diff`  column contains absolute difference between the two factors' percent influence on unadjusted SIOM, not unstandardized SIOM. This means that these calculated absolute differences do not capture the percent influence of the two factors after the USG% and Pace adjustments executed five cells prior to this one, which were applied to `siom_self_factor` and `siom_others_factor`, respectively. Because of this, we had to observe the effect that various possible `siom_others_factor` weights had on `unstded_SIOM`.

Using our selected coefficient, we've now calculated `unstded_SIOM` using the exact formulas presented in the cell where it was calculated. Note that the number at the 92nd position of the Series `possible_coeffs` is equal to the weight on `siom_others_factor` five cells above this one. It had be hard coded in due to the order that our cells were executed.

Now, in the below cell, we will standardize SIOM by season. This is important as our statistic is a measure of player skill, which is fundamentally relative to the other athletes they are competing against.

```python
# Since this metric is seasonal, let's make it standardized by season...
scaler = StandardScaler()
r_siom = pd.Series([], dtype = float)
for name, group in nba_groups:
    arr0 = (group['unstded_SIOM'].values).reshape(-1,1)
    r_siom = r_siom.append(pd.Series(scaler.fit_transform(arr0).flatten()))
r_siom = r_siom.reset_index()
r_siom = pd.Series(r_siom.iloc[:,1])
nba['rSIOM'] = r_siom
```

## Evaluation of Significance

### I. Permutation Test for p-value for Team Offensive Rating vs. the Maximum rSIOM per Team per Season Regression Slope

In the following cells, we use permuted data to determine the significance of the relationship in our data between a team's offensive rating and that team's player with the highest `rSIOM` score on the team. By permuting the data, we are able to generate the circumstances that would be true under the null hypothesis that there is no effect of a team's best offensive engine on their offensive performance.

```python
def permute(series):
    perm = series.copy().values
    np.random.shuffle(perm)
    return pd.Series(perm)

def plot_line(df, model, is_resampled=True):
    if is_resampled:
        color = "grey"
        alpha = 0.1
        label = None
    else:
        color = "darkorange"
        alpha = 1
        label = 'True Slope'
    plt.plot(df['rSIOM'], df['rSIOM'] * model.coef_[0] + model.intercept_, color=color, alpha=alpha, label=label)
```

```python
# Plot true data
plt.scatter(
    best_r_siom['rSIOM'],
    best_r_siom['ortg'],
    alpha = 0.2
)

# 100 iterations of models on permuted data
for i in range(100):
    permuted_model = LinearRegression().fit(best_r_siom[['rSIOM']], permute(best_r_siom['ortg']))
    plot_line(best_r_siom, permuted_model)


# Model true data and plot
plot_line(best_r_siom, lin2, is_resampled=False)
plt.title('True Relationship vs. Null')
plt.xlabel('Player rSIOM')
plt.legend()
plt.ylabel('Team ORTG')
plt.show()
```

In the three cells below, we evaluate the significance of this reliationship by finding the percent of slopes under the null hypothesis that are greater than the one found in our data set. This is the p-value of the linear regression model.

```python
# What fraction of permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
n_perms = 1000
perm_slopes = np.zeros(n_perms)

for i in range(n_perms):
    perm_model = LinearRegression().fit(best_r_siom[['rSIOM']], permute(best_r_siom['ortg']))
    perm_slopes[i] = perm_model.coef_[0]
    
    if np.abs(perm_model.coef_[0]) > np.abs(lin2.coef_):
        steeper_slopes += 1
print('Percent of slopes steeper than that of our model: {:.4%}'.format(steeper_slopes/n_perms))
```

```python
# Plot of our slope in relation to the distribution of slopes expected under the null hypothesis
binsizes, _, _ = plt.hist(perm_slopes, bins=50)
plt.vlines(lin2.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.legend(['Observed Slope','Slopes under Null'])
plt.title('True Slope vs. Slopes under Null')
plt.xlabel('Coefficient of Regression Model')
plt.ylabel('Frequency')
plt.show()
```

```python
# Regression table
sm_model = smf.ols("ortg ~ rSIOM", data= best_r_siom)
result = sm_model.fit()

# P-values
print('Analytic p-value: {:.4f}'.format(result.pvalues.rSIOM))
print('Simulated p-value: {:.4f}'.format(steeper_slopes/n_perms))
```

### II. Confidence Interval for the Mean $f_1$-Score of Our Multi-Class Logistic Classifier

In the following cells, we use bootstrapping to find a 99% confidence interval for the mean $f_1$-score of our logistic classifier. By utilizing bootstrapping, we are able to perform our classification many, many times using datasets similar to ours. This allows us to get a range of realistic possibilities for our mean $f_1$-score.

```python
# Perform many bootstraps...
num_bootstrap = 500
bootstrap_meanf1s = np.zeros(500)
for i in range(num_bootstrap):
    resampled_nba = nba.sample(n=len(nba), replace=True)
    Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    resampled_nba[['rSIOM']], 
    resampled_nba['pos'], 
    test_size= 0.25)
    resampled_logistic = lm.fit(Xr_train, yr_train)
    yr_true = resampled_nba.pos
    yr_pred_full = logistic.predict(resampled_nba[['rSIOM']])
    bootstrap_meanf1s[i] = f1_score(yr_true, yr_pred_full, average = 'micro')
```

```python
# Get simulation-derived estimate of 99% confidence interval for mean f_1 score
sim_est_99pct_ci = np.quantile(bootstrap_meanf1s, q=[0.005, 0.995])
print('Simulated 99% Confidence Interval for f1 Score: ({:.4f},{:.4f})'.format(sim_est_99pct_ci[0], sim_est_99pct_ci[1]))
```

```python
# Plot observed mean f1 score and histogram of resampled mean f1 scores
plt.figure()
binsizes, _, _ = plt.hist(bootstrap_meanf1s, bins=20, color = 'red')
plt.vlines(obs_f1score, ymin=0, ymax=np.max(binsizes), color = 'green')
plt.vlines(sim_est_99pct_ci[0], ymin=0, ymax=np.max(binsizes), color ='black')
plt.vlines(sim_est_99pct_ci[1], ymin=0, ymax=np.max(binsizes), color ='black')
plt.vlines(baseline_guess_rate, ymin=0, ymax=np.max(binsizes), color = 'blue')
plt.title('Distribution of Bootstrap Mean f1 Scores')
plt.legend(['Observed f1 Score','Confidence Interval Lower Bound','Confidence Interval Upper Bound', 
            'Baseline Guess Rate', 'Bootstrap f1 Scores'], fontsize = 'x-small')
plt.show()
```

### I. Interpretations

#### A. Multi-Class Logistic Classifier   
Looking at our multi-class logistic classifier, the overall $f_1$-score is in-fact better than the baseline accuracy. Furthermore, as can be seen by our 99% confidence interval for the $f_1$-score, which we found using bootstrapping, we are 99% confident that our $f_1$-score is statistically significant when compared to the baseline accuracy. While the difference in predictive power is statistically significant, it may not be tremendously meaningful, based solely on the magnitude of this difference. This could be improved with further optimization of the coefficients on the parameters of the statistic. 
    
Another interesting result is the stark difference in $f_1$-scores across the three positional groups. While the guard and center $f_1$-scores are near the overall $f_1$-score, with the guard one being the best, the forward $f_1$-score was incredibly low. There are a handful of reasons why forwards may have been tricky for the classifier to handle. First, looking at our statistic's top-ranking and bottom-ranking seasons, the top-ranking seasons are almost all guards, while the bottom-ranking ones are almost all centers. Additionally, the forward position in the NBA has changed drastically over the past 20 seasons, and has reached a point where it is now more diverse than ever, meaning that the on-court performance and responsibilities of forwards, and consequently their performance metrics, are not as uniform as they used to be.
    
#### B. Linear Regression Models
We used a multivariate linear regression model to determine the impact of certain performance metrics of players on a more broad measure of their offensive productivity, OBPM. The metrics we used in this model were the number of points they scored that were unassisted, which they created themselves, the number of assist points, and their TS%. We performed this regression analysis to gain a better sense of the impact of these variables on a player's offensive contribution and productivity to inform the creation of our own statistic, rSIOM. The score of 0.6722 means that the model explains 67.22% of the variation in obpm, meaing that these 3 variables are decent predictors of a player's OBPM.

In the linear regression model where we regressed a team's offensive rating on the player with the highest value of our statistic, we found a there to be a statistically significant impact of rSIOM on the team's offensive rating. This relationship estimates the effect of a team's best player on its overall offensive rating, which was a goal of our statistic. We used permutation tests to assess whether or not the estimated effect we observed in the dataset was statistically significant, meaning that we wanted to see whether the observed result was due to random chance or due to an actual underlying relationship in the data. Both measurements of the p-value yielded 0. The analytic p-value was calculated using the standard linear regression model, whereas the simulated p-value was calculated based off of the relationships found in the permuted data. By permuting the data, we can create a distribution of slopes that we could expect to see under the null hypothesis that there is no relationship between a team's best player, based on their rSIOM value, and their offensive rating. This allowed us to ultimately compare the slopes under the null the the slope of the observed relationship, which proved it to be statistically significant. This analysis proves that there is a significant impact of a team's most valuable offensive player on the team's overall offensive rating.

Examining the rSIOM values for various players from different seasons, our statistic seems to be a reasonable representation of a player's ability to induce and convert offensive opportunities for himself, as well as leading teammates to those opportunities.
