### I. Creating Our Statistic

In this portion of our project, we create our statistic, `si_offense`.

```python
# Creating a new df to calculate and store the expected value of shots from 0-3 ft, 3-10 ft, 10-16 ft, 
# 16 ft to the 3PT line, corner 3s, and arc 3s for each individual season...
expected_values_by_loc = lg_avg_shooting.copy()
# Need to find % 3PA from arc (non-corner) and Arc 3PM %
expected_values_by_loc['pct_3pa_arc'] = 1 - expected_values_by_loc['pct_3pa_corner']
# Now, using the fact that overall 3PM % is a weighted average of arc and corner 3PM %, we know that...
expected_values_by_loc['pct_3pm_arc'] = round(((expected_values_by_loc['pct_3pm'] - 
                                        (expected_values_by_loc['pct_3pm_corner']*
                                         expected_values_by_loc['pct_3pa_corner']))/expected_values_by_loc['pct_3pa_arc']),3)
expected_values_by_loc.head(5)
```

In the above cell, we created a new DataFrame expected_values_by_loc which will contain the expected value of shots from various locations on the floor from 2000 to 2021. Furthermore, we created two columns which were not present in our read-in data `pct_3pa_arc` and `pct_3pm_arc`, which contains the league average percentage of 3s taken that were not in the corner and the conversion rate of non-corner 3s by season, respectively.

```python
# Now, finding the expected values...
expected_values_by_loc['ev_0to3ft'] = 2*(expected_values_by_loc['pct_fgm_0to3_ft'])
expected_values_by_loc['ev_3to10ft'] = 2*(expected_values_by_loc['pct_fgm_3to10_ft'])
expected_values_by_loc['ev_10to16ft'] = 2*(expected_values_by_loc['pct_fgm_10to16_ft'])
expected_values_by_loc['ev_16ftto3pt'] = 2*(expected_values_by_loc['pct_fgm_16ftto3pt'])
expected_values_by_loc['ev_3pt_corner'] = 3*(expected_values_by_loc['pct_3pm_corner'])
expected_values_by_loc['ev_3pt_arc'] = 3*(expected_values_by_loc['pct_3pm_arc'])
expected_values_by_loc

# Removing all unnecessary columns from expected_values_by_loc...
expected_values_by_loc = expected_values_by_loc.loc[:,['ev_0to3ft', 'ev_3to10ft', 'ev_10to16ft',
                                                       'ev_16ftto3pt', 'ev_3pt_corner', 'ev_3pt_arc']]
expected_values_by_loc.head(5)
```

In the above cell, we calculated and put into columns the expected value of shots from six different areas on the court for every season in our dataset (2000-2001 through 2020-2021) and then removed unnecessary columns from our DataFrame which contains the aforementioned expecrted values.

Note that pbpstats classifies short mid-range shots as shots from 4-14 ft away from the basket and long mid-range as coming from 14 ft away from the basket to the three point line. Because of how the Basketball Reference league average shooting data is broken up, we will have to define short mid-range as 3-16 ft away and long mid-range as 16 ft away to the three point line. Looking at the percentages in our lg_avg_shooting DataFrame, this re-classification should not alter our results significantly. In the below cell, we adjust our expected_values_by_loc DataFrame.

```python
# Adjusting for reclassification
for i in list(lg_avg_shooting.index.values):
    expected_values_by_loc.loc[i,'estev_3to16ft'] = round((((lg_avg_shooting.loc[i,'pct_fga_3to10_ft']/(lg_avg_shooting.loc[i,'pct_fga_3to10_ft'] 
                                                       + lg_avg_shooting.loc[i,'pct_fga_10to16_ft']))*expected_values_by_loc.loc[i,'ev_3to10ft']) 
                                                       + ((lg_avg_shooting.loc[i,'pct_fga_10to16_ft']/(lg_avg_shooting.loc[i,'pct_fga_3to10_ft'] 
                                                       + lg_avg_shooting.loc[i,'pct_fga_10to16_ft']))*expected_values_by_loc.loc[i,'ev_10to16ft'])),3)
expected_values_by_loc.head(5)
```

In the below cell, we find the `si_offense` values for each player season present in our dataset. `si_offense` is made up of two sub-metrics we've calculated, `si_self_factor` and `si_others_factor`. `si_self_factor` is a measure of a player's self-induced scoring, while `si_others_factor` is a measure of a player's playmaking. Both factors take into account both efficiency and production.

Additionally, after summing both factors, two further adjustments are made before getting our final unstandardized `si_offense` values. These adjustments are for player Usage Rate (`usg%`) and for player Pace (`pace_estimate`). This is to prevent our statistic from favoring players who use a higher proportion of their team possesions while on the floor (which is measured by Usage Rate) or who have a larger number of possesions to produce points (which is measured by Pace). This is because while production is significant, our statistic aims to measure a player's ability to induce offense, which is not the same as capturing a player's production while in that role.

Note that in the above output, there was no weight on either `si_self_factor` or `si_others_factor` when computing `unstded_si_offense`. By the output above, without performing further analysis, we can tell that `si_others_factor` must have a disproportionate influence on any given player's `unstded_si_offense`. So, in the next two cells, we will attempt to find the optimal weight for `si_others_factor` before standardizing our statistic by season. The new weight will be used in the cell where `unstded_si_offense` is calculated, however, the above output was obtained utilizing no weight on `si_others_factor`.

```python
# Finding optimal weight for "si_others_factor"...
possible_coeffs = list(np.linspace(0.01,1,1000))
pct_influence = pd.DataFrame([pd.Series(np.zeros(1000)), pd.Series(np.zeros(1000))],
                             columns = ['si_self', 'si_others'],
                             dtype = float)
summed_si_self_factor = sum(nba['si_self_factor'])
for i in range(0,1000):
    alt_si_others_factor = possible_coeffs[i]*nba['si_others_factor']
    summed_alt_si_others_factor = sum(alt_si_others_factor)
    summed_unadj_si = summed_si_self_factor + summed_alt_si_others_factor
    pct_influence.loc[i,'si_self'] = summed_si_self_factor/summed_unadj_si
    pct_influence.loc[i,'si_others'] = summed_alt_si_others_factor/summed_unadj_si
pct_influence
```

In the cell above, we found the percent influence of `si_self_factor` and `si_others_factor` when multiplying `si_others_factor` by 1000 different possible coefficients before adding the two to retrive resultant unadjusted `unstded_si_offense` values.

Now, in the cell below, we will find the absolute difference between the percent influence of both factors and find the sensible weights to place on `si_others_factor`.

```python
pct_influence['diff'] = abs(pct_influence['si_self'] - pct_influence['si_others'])
print(pct_influence.describe())
print(pct_influence[pct_influence['diff'] <= .6])
```

After some trial and error, we settled on the above coefficient. Using it, we've now calculated `unstded_si_offense` using the exact formulas presented in the cell where the metric was found.

#### A. Multi-Class Logistic Classificaiton of Player Position ('pos') on Our Statistic ('si_offense')

Here, we use a multi-class logistic regression model to attempt to classify a player's position given their `si_offense` value. We do this to see if our statistic is biased in any way, namely by over- or under-predicting a certain position. In this first cell, we divide our data into training and testing sets so that we have some data on which to fit the model and some unseen data on which to test its predictive power.

```python
# train-test split (75%/25%)
Xl_train, Xl_test, yl_train, yl_test = train_test_split(
    nba[['si_offense']], 
    nba['pos'], 
    test_size= 0.25
)
print('Length of training data set: {} \nLength of training data set:'.format(len(yl_train)),len(yl_test))
```

```python
# Setting up logistic model
lm = LogisticRegression(multi_class = 'multinomial', solver = 'sag')
logistic = lm.fit(Xl_train, yl_train)
lm_score = logistic.score(Xl_test, yl_test)
print('Score of fitting the model on one instance of training and testing data subsets: {: .4f}'.format(lm_score))
```

In the cell above, we use the training and testing subsets of the data that we created to fit the logistic regression model. We then test that fitted model on data that the model has not yet seen in order to see its predictive capabilities. 

In the cell below, we use cross-validation to repeat that same process 30 times with varying training and testing subsets to achieve a better expected accuracy from the model. 

```python
scores_log = cross_val_score(lm,nba[['si_offense']],nba['pos'],cv=30)
print('Mean cross-validated accuracy: {:.4f}.'.format(scores_log.mean()))
```

## Evaluation of Significance

### I. Permutation Test for p-value for Team Offensive Rating vs. the Maximum of Our Statistic per Team per Season Regression Slope

In the following cells, we use permuted data to determine the significance of the relationship in our data between a team's offensive rating and that team's player with the highest `si_offense` score on the team. By permuting the data, we are able to generate the circumstances that would be true under the null hypothesis that there is no effect of a team's best offensive engine on their offensive performance.

```python
def permute(series):
    perm = series.copy().values
    np.random.shuffle(perm)
    return pd.Series(perm)

def plot_line(df, model, is_resampled=True):
    if is_resampled:
        color = "grey"
        alpha = 0.1
        label = None
    else:
        color = "darkorange"
        alpha = 1
        label = 'True Slope'
    plt.plot(df['si_offense'], df['si_offense'] * model.coef_[0] + model.intercept_, color=color, alpha=alpha, label=label)
```

```python
# Plot true data
plt.scatter(
    best_si['si_offense'],
    best_si['ortg'],
    alpha = 0.2
)

# 100 iterations of models on permuted data
for i in range(100):
    permuted_model = LinearRegression().fit(best_si[['si_offense']], permute(best_si['ortg']))
    plot_line(best_si, permuted_model)


# Model true data and plot
plot_line(best_si, lin2, is_resampled=False)
plt.title('True Relationship vs. Null')
plt.xlabel('si_offense')
plt.legend()
plt.ylabel('Team ORTG')
plt.show()
```

In the 3 cells below, we evaluate the significance of this reliationship by finding the percent of slopes under the null hypothesis that are greater than the one found in our data set. This is the p-value of the linear regression model.

```python
# What fraction of permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
n_perms = 1000
perm_slopes = np.zeros(n_perms)

for i in range(n_perms):
    perm_model = LinearRegression().fit(best_si[['si_offense']], permute(best_si['ortg']))
    perm_slopes[i] = perm_model.coef_[0]
    
    if np.abs(perm_model.coef_[0]) > np.abs(lin2.coef_):
        steeper_slopes += 1
print('Percent of slopes steeper than that of our model: {:.4%}.'.format(steeper_slopes/n_perms))
```

```python
# Plot of our slope in relation to the distribution of slopes ewxpected under the null hypothesis
binsizes, _, _ = plt.hist(perm_slopes, bins=50)
plt.vlines(lin2.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.legend(['Observed Slope','Slopes under Null'])
plt.title('True Slope vs. Slopes under Null')
plt.xlabel('Coefficient of Regression Model')
plt.ylabel('Frequency')
plt.show()
```

```python
# Regression table
sm_model = smf.ols("ortg ~ si_offense", data=best_si)
result = sm_model.fit()

# P-values
print('Analytic p-value: {:.4f}.'.format(result.pvalues.si_offense))
print('Simulated p-value: {:.4f}.'.format(steeper_slopes/n_perms))
```

### II. Confidence Interval for the Mean $f_1$-Score of Our Multi-Class Logistic Classifier

In the following cells, we use bootstrapping to find a 99% confidence interval for the mean $f_1$-score of our logistic classifier. By utilizing bootstrapping, we are able to perform our classification many, many times using datasets similar to ours. This allows us to get a range of realistic possibilities for our mean $f_1$-score. 

```python
# Perform many bootstraps...
num_bootstrap = 500
bootstrap_meanf1s = np.zeros(500)
for i in range(num_bootstrap):
    resampled_nba = nba.sample(n=len(nba), replace=True)
    Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    resampled_nba[['si_offense']], 
    resampled_nba['pos'], 
    test_size= 0.25)
    resampled_logistic = lm.fit(Xr_train, yr_train)
    yr_true = resampled_nba.pos
    yr_pred_full = logistic.predict(resampled_nba[['si_offense']])
    bootstrap_meanf1s[i] = f1_score(yr_true, yr_pred_full, average = 'micro')
```

```python
# Get simulation-derived estimate of 99% confidence interval for mean f_1 score
sim_est_99pct_ci = np.quantile(bootstrap_meanf1s, q=[0.005, 0.995])
print('Simulated 99% Confidence Interval for f1 Score: ({:.4f},{:.4f})'.format(sim_est_99pct_ci[0], sim_est_99pct_ci[1]))
```

```python
# plot observed mean f1 score and histogram of resampled mean f1 scores
plt.figure()
binsizes, _, _ = plt.hist(bootstrap_meanf1s, bins=20, color = 'red')
plt.vlines(obs_f1score, ymin=0, ymax=np.max(binsizes), color = 'green')
plt.vlines(sim_est_99pct_ci[0], ymin=0, ymax=np.max(binsizes), color ='black')
plt.vlines(sim_est_99pct_ci[1], ymin=0, ymax=np.max(binsizes), color ='black')
plt.vlines(baseline_guess_rate, ymin=0, ymax=np.max(binsizes), color = 'blue')
plt.title('Distribution of Bootstrap f1 Scores')
plt.legend(['Observed f1 Score','Confidence Interval Lower Bound','Confidence Interval Upper Bound', 
            'Baseline Guess Rate', 'Bootstrap f1 Scores'], fontsize = 'x-small')
plt.show()
```

## Interpretation and Conclusions

Examining the si_offense values for various players from different seasons, our statistic seems to be a reasonable representation of a player's ability to induce and convert offensive opportunities for himself, as well as leading teammates to those opportunities. However, the weight of each of its two factors, as well as the weights of the components of these individual factors, could potentially be further altered than done here to produce a slightly better metric. Our stat seems to greatly reward efficient scoring, and low turnover, efficient shot creation, as it should. However, it may favor low turnover rates and penalize high usage rates too much. This can be seen looking at the top-ranked player seasons.
    
Looking at our multi-class logistic classifier, the overall f1 score is in-fact better than the baseline accuracy. Furthermore, as can be seen by our 99% confidence interval for the f1 score, which we found using bootstrapping, we are 99% confident that our f1 score is statistically significant when compared to the baseline accuracy. This is because the baseline accuracy was not apart of our confidence interval. Thus, our classifier does predict player position via our statistic significantly better than the baseline method does.
    
However, another interesting result is the stark difference in f1 scores across the 3 positional groups. While the guard and center f1 scores are near the overall f1 score, with the guard one being the best, the forward f1 score was incredibly low. There a handful of reasons why forwards may have been this tricky for the classifier to handle. First, looking at our statistic's top-ranking and bottom-ranking seasons, the top-ranking seasons are almost all guards, while the bottom-ranking ones are almost all centers. Additionally, the forward position in the NBA has changed drastically over the past 20 seasons, and has reached a point where it is now more diverse than ever.
    
We used a multivariate linear regression model to determine the impact of certain performance metrics of the players on a more broad measure of their offensive productivity, offensive box plus/minus (obpm). The metrics we used in this model were the number of points they scored that were unassisted, which they created themselves, the number of assist points, and their true shooting percentage. The obpm of a player measures their contribution to the team when they are on versus off the court. We estimated this regression to gain a better sense of the impact of these variables on a player's offensive contribution and productivity to inform the creation of our own statistic, si_offense. The score of 0.6722 means that the model explains 67.22% of the variation in obpm, meaing that these 3 variables are decent predictors of a player's obpm.
    
In the linear regression model where we regressed a team's offensive rating on the player with the highest value of our statistic, we found a there to be a statistically significant impact of si_offense on the team's ortg. This relationship estimates the effect of a team's best player on its overall offensive rating, which was a goal of our statistic. We used permutation tests to assess whether or not the estimated effect we observed in the dataset was statistically significant, meaning that we wanted to see whether the observed result was due to random chance or due to an actual underlying relationship in the data. Both measurements of the p-value yielded 0. The analytic p-value was calculated using the standard linear regression model, whereas the simulated p-value was calculated based off of the relationships found in the permuted data. By permuting the data, we can create a distribution of slopes that we could expect to see under the null hypothesis that there is no relationship between a team's best player, based on their si_offense value, and their offensive rating (ortg). This allowed us to ultimately compare the slopes under the null the the slope of the observed relationship, which proved it to be statistically significant. This analysis proves that there is a significant impact of a team's most valuable offensive player on the team's overall offensive rating.
