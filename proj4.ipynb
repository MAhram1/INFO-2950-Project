### I. Creating Our Statistic

```python
# Creating a new df to calculate and store the expected value of shots from 0-3 ft, 3-10 ft, 10-16 ft, 
# 16 ft to the 3PT line, corner 3s, and arc 3s for each individual season...
expected_values_by_loc = lg_avg_shooting.copy()
# Need to find % 3PA from arc (non-corner) and Arc 3PM %
expected_values_by_loc['pct_3pa_arc'] = 1 - expected_values_by_loc['pct_3pa_corner']
# Now, using the fact that overall 3PM % is a weighted average of arc and corner 3PM %, we know that...
expected_values_by_loc['pct_3pm_arc'] = round(((expected_values_by_loc['pct_3pm'] - 
                                        (expected_values_by_loc['pct_3pm_corner']*
                                         expected_values_by_loc['pct_3pa_corner']))/expected_values_by_loc['pct_3pa_arc']),3)
expected_values_by_loc
```

In the above cell, we created a new DataFrame expected_values_by_loc which will contain the expected value of shots from various locations on the floor for each season in our dataset. Furthermore, we created two columns which were not present in our read-in data "pct_3pa_arc" and "pct_3pm_arc", which contains the league average percentage of 3s taken that were not in the corner and the conversion rate of non-corner 3s by season.

```python
# Now, finding the expected values...
expected_values_by_loc['ev_0to3ft'] = 2*(expected_values_by_loc['pct_fgm_0to3_ft'])
expected_values_by_loc['ev_3to10ft'] = 2*(expected_values_by_loc['pct_fgm_3to10_ft'])
expected_values_by_loc['ev_10to16ft'] = 2*(expected_values_by_loc['pct_fgm_10to16_ft'])
expected_values_by_loc['ev_16ftto3pt'] = 2*(expected_values_by_loc['pct_fgm_16ftto3pt'])
expected_values_by_loc['ev_3pt_corner'] = 3*(expected_values_by_loc['pct_3pm_corner'])
expected_values_by_loc['ev_3pt_arc'] = 3*(expected_values_by_loc['pct_3pm_arc'])
expected_values_by_loc
```

```python
# Removing all unnecessary, extraneous columns from expected_values_by_loc...
expected_values_by_loc = expected_values_by_loc.loc[:,['ev_0to3ft', 'ev_3to10ft', 'ev_10to16ft',
                                                       'ev_16ftto3pt', 'ev_3pt_corner', 'ev_3pt_arc']]
expected_values_by_loc
```

In the above two cells, we calculated and put into columns the expected value of shots from six different areas on the court for every season in our dataset (2000-2001 through 2020-2021) and then removed unnecessary columns from our DataFrame which contains the aforementioned expecrted values.

Note that pbpstats classifies short mid-range shots as shots from 4-14 ft away from the basket and long mid-range as coming from 14 ft away from the basket to the three point line. Because of how the Basketball Reference league average shooting data is broken up, we will have to define short mid-range as shots coming from 3-16 ft away and long mid-range as coming from 16 ft away to the three point line. Looking at the percentages in our lg_avg_shooting DataFrame, this re-classification should not alter our results significantly.

```python
# Adjusting for reclassification
for i in list(lg_avg_shooting.index.values):
    expected_values_by_loc.loc[i,'estev_3to16ft'] = round((((lg_avg_shooting.loc[i,'pct_fga_3to10_ft']/(lg_avg_shooting.loc[i,'pct_fga_3to10_ft'] 
                                                       + lg_avg_shooting.loc[i,'pct_fga_10to16_ft']))*expected_values_by_loc.loc[i,'ev_3to10ft']) 
                                                       + ((lg_avg_shooting.loc[i,'pct_fga_10to16_ft']/(lg_avg_shooting.loc[i,'pct_fga_3to10_ft'] 
                                                       + lg_avg_shooting.loc[i,'pct_fga_10to16_ft']))*expected_values_by_loc.loc[i,'ev_10to16ft'])),3)
expected_values_by_loc
```

```python
# Piece-by-piece building of our statistic
nba_groups = nba.groupby('season')
si_self_factor = pd.Series([], dtype = float)
si_others_factor = pd.Series([], dtype = float)
for name, group in nba_groups:
    si_self_factor = si_self_factor.append(((1 - (group['shotqualityavg']))*(group['efg%'] - lg_avg_teamstats.loc[name, 'efg%'])
                            *(group['ptsunassisted2s'] + group['ptsunassisted3s'])) + (group['ft%'] - 
                            (lg_avg_teamstats.loc[name, 'ft'])/lg_avg_teamstats.loc[name, 'fta'])*group['ftpoints'])
    si_others_factor = si_others_factor.append((1.5 - (group['tov%']/lg_avg_teamstats.loc[name, 'tov%']))
                                *((group['atrimassists']*(expected_values_by_loc.loc[name,'ev_0to3ft']))
                                + (group['shortmidrangeassists']*(expected_values_by_loc.loc[name,'estev_3to16ft']))
                                + (group['longmidrangeassists']*(expected_values_by_loc.loc[name,'ev_16ftto3pt']))
                                + (group['corner3assists']*(expected_values_by_loc.loc[name,'ev_3pt_corner']))
                                + (group['arc3assists']*(expected_values_by_loc.loc[name,'ev_3pt_arc']))))
nba['si_self_factor'] = si_self_factor
nba['si_others_factor'] = (1/10)*si_others_factor
nba['si_unadjusted'] = nba['si_self_factor'] + nba['si_others_factor']
# Adjusting for player USG%..
nba['uSI_offense'] = (1 - np.log(nba['usg%']/20))*nba['si_unadjusted']
# Adjusting for pace...
nba['pace_estimate'] = (48*nba['offposs'])/nba['mp']
nba_groups = nba.groupby('season')
unstded_si_offense = pd.Series([], dtype = float)
for name, group in nba_groups:
    unstded_si_offense = unstded_si_offense.append((1-np.log(group['pace_estimate']/lg_avg_teamstats.loc[name,'pace']))*group['uSI_offense'])
nba['unstded_si_offense'] = unstded_si_offense   
```

```python
nba.sort_values(by = 'unstded_si_offense', ascending = False)
```

Note that in the above output, there was no weight on either "si_self_factor" or "si_others_factor". By the output above, without performing further analysis, we can tell that "si_others_factor" must have a disproportionate influence on any given player's si_offense. So, in the next code cell, we will attempt to find the optimal weight for "si_others_factor" before standardizing our statistic by season. The new weight will be used in the cell where "unstded_si_offense" is calculated, however, the above output was obtained utilizing no weight on "si_others_factor".

```python
# Finding optimal weight for "si_others_factor"...
possible_coeffs = list(np.linspace(0.01,1,1000))
pct_influence = pd.DataFrame([pd.Series(np.zeros(1000)), pd.Series(np.zeros(1000))],
                             columns = ['si_self', 'si_others'],
                             dtype = float)
summed_si_self_factor = sum(nba['si_self_factor'])
for i in range(0,1000):
    alt_si_others_factor = possible_coeffs[i]*nba['si_others_factor']
    summed_alt_si_others_factor = sum(alt_si_others_factor)
    summed_unadj_si = summed_si_self_factor + summed_alt_si_others_factor
    pct_influence.loc[i,'si_self'] = summed_si_self_factor/summed_unadj_si
    pct_influence.loc[i,'si_others'] = summed_alt_si_others_factor/summed_unadj_si
pct_influence
```

```python
pct_influence['diff'] = abs(pct_influence['si_self'] - pct_influence['si_others'])
print(pct_influence.describe())
print(pct_influence[pct_influence['diff'] <= .6])
possible_coeffs[91]
```

```python
# Since this metric is seasonal, let's make it standardized by season...
scaler = StandardScaler()
si_offense = pd.Series([], dtype = float)
for name, group in nba_groups:
    arr0 = (group['unstded_si_offense'].values).reshape(-1,1)
    si_offense = si_offense.append(pd.Series(scaler.fit_transform(arr0).flatten()))
si_offense = si_offense.reset_index()
si_offense = pd.Series(si_offense.iloc[:,1])
nba['si_offense'] = si_offense
nba
```

### II. Preregistered Analyses 

#### A. Multi-Class Logistic Classificaiton of Player Position ('pos') on Our Statistic ('si_offense')

```python
# train-test split (75%/25%)
Xl_train, Xl_test, yl_train, yl_test = train_test_split(
    nba[['si_offense']], 
    nba['pos'], 
    test_size= 0.25
)
print("Length of train and test sets:", len(yl_train), len(yl_test))
```

```python
# Setting up logistic model
lm = LogisticRegression(multi_class = 'multinomial', solver = 'sag')
logistic = lm.fit(Xl_train, yl_train)
logistic.score(Xl_test, yl_test)
```

```python
scores_log = cross_val_score(lm,nba[['si_offense']],nba['pos'],cv=30)
print(scores_log)
print(f"Mean cross-validated accuracy: {scores_log.mean() : .4f}")
```

```python
# Baseline guess of most common position (G)
baseline_guess_rate = (sum(nba.pos == "G"))/len(nba)
print(f'Baseline guess rate: {(sum(nba.pos == "G")/len(nba)) : .4f}')
```

```python
# True labels and predicted labels
yl_true = nba.pos
yl_pred_full = logistic.predict(nba[['si_offense']])

# Confusion Matrix
disp = ConfusionMatrixDisplay(
    confusion_matrix(yl_true, yl_pred_full), 
    display_labels = nba.pos.unique()
)
disp.plot()
plt.title('Confusion Matrix')
plt.xticks(ticks = [0,1,2], labels = ['C','F','G'])
plt.yticks(ticks = [0,1,2], labels = ['C','F','G'])
plt.show()
```

```python
# Classification Report
print(classification_report(yl_true, yl_pred_full))
obs_f1score = f1_score(yl_true, yl_pred_full, average = 'micro')
print(obs_f1score)
```

#### B. Multivariate Regression of OBPM (obpm) With Unassisted Points (ptsunassisted), Assist Points (assistpoints), and True Shooting Percentage (ts%) as the Input Variables

```python
# train-test split (75%/25%)
nba['ptsunassisted'] = nba.ptsunassisted2s + nba.ptsunassisted3s
Xm_train, Xm_test, ym_train, ym_test = train_test_split(
    nba[['ptsunassisted','assistpoints','ts%']], 
    nba['obpm'], 
    test_size= 0.25
)
print("Length of train and test sets:", len(ym_train), len(ym_test))
```

```python
# Fitting, predicting, and plotting the results of the multivariate linear model
multi_lr = LinearRegression()
multi_lin = multi_lr.fit(Xm_train,ym_train)
ym_pred = multi_lin.predict(Xm_test)

# Plotting predicted offensive box plus/minus vs. actual bpm
plt.scatter(
    ym_test,
    ym_pred,
    alpha = 0.3
)
plt.plot(
    np.linspace(-20,20),
    np.linspace(-20,20),
    linestyle='dashed',
    linewidth=2,
    color = 'black',
)
plt.show()
```

```python
scores_multi = cross_val_score(multi_lr, nba[['ptsunassisted','assistpoints','ts%']], nba['obpm'],cv=30)
print(scores_multi)
print(f"Mean cross-validated accuracy: {scores_multi.mean() : .4f}")
```

### I. Permutation Test for p-value for Team Offensive Rating vs. the Maximum of Our Statistic per Team per Season Regression Slope

```python
def permute(series):
    perm = series.copy().values
    np.random.shuffle(perm)
    return pd.Series(perm)

def plot_line(df, model, is_resampled=True):
    if is_resampled:
        color="grey"
        alpha=0.1
    else:
        color="darkorange"
        alpha=1
    plt.plot(df['si_offense'], df['si_offense'] * model.coef_[0] + model.intercept_, color=color, alpha=alpha)
```

```python
# Plot true data
plt.scatter(
    best_si['si_offense'],
    best_si['ortg'],
    alpha = 0.2
)

# 100 iterations of models on permuted data
for i in range(100):
    permuted_model = LinearRegression().fit(best_si[['si_offense']], permute(best_si['ortg']))
    plot_line(best_si, permuted_model)

# Model true data and plot
plot_line(best_si, lin2, is_resampled=False)
plt.title('True Relationship vs. Null')
plt.show()
```

```python
# What fraction of permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
n_perms = 1000
perm_slopes = np.zeros(n_perms)

for i in range(n_perms):
    perm_model = LinearRegression().fit(best_si[['si_offense']], permute(best_si['ortg']))
    perm_slopes[i] = perm_model.coef_[0]
    
    if np.abs(perm_model.coef_[0]) > np.abs(lin2.coef_):
        steeper_slopes += 1
print(f"Percent of slopes steeper than that of our model: {(steeper_slopes / n_perms) : .4%}")
```

```python
# Plot of our slope in relation to the distribution of slopes ewxpected under the null hypothesis
binsizes, _, _ = plt.hist(perm_slopes, bins=50)
plt.vlines(lin2.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.legend(['Observed Slope','Slopes under Null'])
plt.show()
```

## Data Limitations

    While there are not any true potential harms or data gaps, however, a key limitation of our dataset is the aforementioned fact that it only spans the previous 20 NBA seasons (the 2000-2001 season through the 2020-2021 season). This will not be an issue when our statistic based on this dataset is deployed in the previously mentioned context we envision its primary use to be in, which is for making personnel decisions such as signing a free agent or executing a trade, current season player evaluation, and for game planning purposes. But, it will limit our metric’s use in historical comparisons and all-time player evaluation, as it will only be applied to players who played in the 2000-2001 season or later.
